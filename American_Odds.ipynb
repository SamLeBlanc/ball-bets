```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('american_odds.sqlite')
```

```python
query = "SELECT name FROM sqlite_master WHERE type='table';"
tables_df = pd.read_sql_query(query, conn)
```

```python
# Initialize an empty DataFrame to store the concatenated data
all_tables_data = pd.DataFrame()

# Iterate through the table names and read each table into a DataFrame
for index, row in tables_df.iterrows():
    table_name = '"' + row['name'] + '"'
    table_data = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)

    # Add a 'table_name' column to identify the source table of each row
    table_data['table_name'] = table_name

    # Append the table data to the all_tables_data DataFrame
    all_tables_data = all_tables_data.append(table_data, ignore_index=True)

# Close the connection
conn.close()
```

```python
odds_full = all_tables_data[['Date','Home','Away','OU','Spread','ML_Home','ML_Away','Points','Win_Margin','table_name']]
```
